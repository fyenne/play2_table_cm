```python
a = widgets.FloatText(description='a', value = 1)
# b = widgets.FloatText(description='b', value = 1)
# c = widgets.IntSlider(description='c')
def f(a):
    print('{}*loge({})={}'.format(a, a, a*np.log(a)))

out = widgets.interactive_output(f, {'a': a })

widgets.HBox([widgets.VBox([a]), out])
```

```python
x = np.arange(.001, 5, step = .001)
y = np.log(x)*x
```

```python
import plotly.graph_objects as go



fig = go.Figure(data=go.Scatter(x=x, y = y))
fig.add_trace(go.Scatter(x = np.repeat(.5, 15), \
    y = np.arange(-4,3), mode = 'lines', name = '.5') )
fig.add_trace(go.Scatter(x = np.repeat(.16667, 15), \
    y = np.arange(-4,3), mode = 'lines', name = '.16667') )
fig.update_xaxes(range = [-.5,5])
fig.update_yaxes(dtick = 1, range = [-4,2])
fig.show()
```
